```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
loops = 2
petals = 2
rotation = 0 * np.pi
```

```python
# Generating angles from 0 to 2*pi
thetas = np.linspace(0, loops * 2 * np.pi - 1e-10, 5000)
theta_max = np.max(theta)

# Generating amplitudes

amp_1 = np.vectorize(lambda theta: (np.cos(petals * (theta + rotation) + 1)) / 2)(theta)

def get_steps(theta):
    theta // np.pi

steps = (1 + np.floor_divide(thetas, 2 * np.pi)) / loops

# Generating complex numbers on the unit circle
z = amp_1 * np.exp(1j * theta)

# Extracting real and imaginary parts
x = np.real(z)
y = np.imag(z)
```

```python
# Plotting the complex unit circle
plt.figure(figsize=(6, 6))
plt.scatter(x, y)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Complex Unit Circle')
plt.grid(True)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```
