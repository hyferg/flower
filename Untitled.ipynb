```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline
```

```python
loops = 2
petals = 2
rotation = 0 * np.pi
```

```python
squish = 1
```

```python
# Generating angles from 0 to 2*pi
gammas = np.linspace(0, 2 * np.pi / squish, 200)

x = gammas
y = np.sin(gammas*squish)

theta = 1/2 * np.pi

x1 = x * np.cos(theta) - y * np.sin(theta)
y1 = x * np.sin(theta) + y * np.cos(theta) - 2 * np.pi
```

```python
# Plotting the complex unit circle
plt.figure(figsize=(6, 6))
plt.scatter(x1, y1)
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Complex Unit Circle')
plt.grid(True)
# plt.xlim(-1,1)
# plt.ylim(-1,1)
plt.show()
```
